```python
import pandas as pd

# Load the cleaned data
df = pd.read_csv('data/district_monthly_weather_clean.csv')

# Analyze by district
colombo_data = df[df['district'] == 'Colombo']

# Find average temperature by year
yearly_avg = df.groupby('year')['mean_temperature'].mean()

# Plot precipitation trends
import matplotlib.pyplot as plt
df.groupby('year_month')['total_precipitation_hours'].sum().plot()
plt.show()
```
